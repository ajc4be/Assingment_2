```python
import pandas as pd
import numpy as np
import seaborn as sns
# Importing the numpy, pandas, and seaborn packages into the workspace
```

```python
df = pd.read_csv('airbnb_hw.csv', low_memory=False)
print( df.shape, '\n')
df.head()
# Displaying Airbnb data in tabular form
```

```python
price = df['Price']
price.unique()

# Displaying all the prices in order to look for areas to clean
# The values need to be converted to numeric and rid of the commas for values in the thousands
```

```python
price = df['Price']
price = price.str.replace(',','') # Replacing the commas in the thousand values with nothing
print( price.unique() , '\n')
price = pd.to_numeric(price,errors='coerce') # Typecasting price to float/numeric
print( price.unique() , '\n')
print( 'Total missing: ', sum( price.isnull() ) ) # Finally converting all the values
```

#Q2

```python
df = pd.read_csv('sharks.csv', low_memory=False)
df.head()
# Importing sharks.csv and displaying in tabular form
```

```python
df['Type'].value_counts()

# Displaying value counts of each Type
```

```python
type = df['Type'] # Creating temporary vector of values for Type

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # Categorizing into Watercraft related
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # Cleaning values
type.value_counts()

df['Type'] = type # Replacing Type variable with the cleaned version
del type # Destroying temporary vector

df['Type'].value_counts() #Displaying counts in new categories
```

```python
df = pd.read_csv('VirginiaPretrialData2017.csv', low_memory=False)
df.head()

# Importing and displaying data in tabular form
```

```python
release = df['WhetherDefendantWasReleasedPretrial']
print(release.unique(),'\n')
print(release.value_counts(),'\n')
release = release.replace(9,np.nan) # Cleaning 9s because they are unclear
print(release.value_counts(),'\n')
sum(release.isnull())
df['WhetherDefendantWasReleasedPretrial'] = release # Replace data column with cleaned values
del release
```

#Q4

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

length = pd.to_numeric(length,errors='coerce') # Convert to numeric
length_NA = length.isnull() # Create missing dummy
print( np.sum(length_NA),'\n')

print( pd.crosstab(length_NA, type), '\n')

length = length.mask( type == 4, 0) # Replace length with 0 when type ==4
length = length.mask( type == 9, np.nan) # Replace length with np.nan when type == 9

length_NA = length.isnull() # Create a new missing dummy
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')

df['ImposedSentenceAllChargeInContactEvent'] = length # Replace data with cleaned version
del length, type # Delete temporary length/type variables
```
